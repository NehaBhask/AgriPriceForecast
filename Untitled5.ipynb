```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder
import joblib
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('Agriculture_price_dataset.csv')

# Data Cleaning & Preparation
print("Initial data shape:", df.shape)
print("Columns:", df.columns.tolist())

# Convert Price Date to datetime
df['Price Date'] = pd.to_datetime(df['Price Date'], format='mixed', dayfirst=True)

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Drop rows with missing critical values
df = df.dropna(subset=['Modal_Price', 'STATE', 'District Name', 'Market Name', 'Commodity', 'Price Date'])

# Create month and year columns for aggregation
df['Year'] = df['Price Date'].dt.year
df['Month'] = df['Price Date'].dt.month
df['Year-Month'] = df['Price Date'].dt.to_period('M')

# Aggregate to monthly average price per commodity per market
monthly_avg = df.groupby(['STATE', 'District Name', 'Market Name', 'Commodity', 'Year', 'Month', 'Year-Month'])['Modal_Price'].mean().reset_index()
monthly_avg = monthly_avg.sort_values(['STATE', 'District Name', 'Market Name', 'Commodity', 'Year', 'Month'])

# Feature Engineering Function
def create_features(df, group_cols):
    """
    Create time-series features for forecasting
    """
    features_df = df.copy()

    # Create lag features
    features_df['lag_1'] = features_df.groupby(group_cols)['Modal_Price'].shift(1)
    features_df['lag_2'] = features_df.groupby(group_cols)['Modal_Price'].shift(2)
    features_df['lag_3'] = features_df.groupby(group_cols)['Modal_Price'].shift(3)

    # Create rolling mean features
    features_df['rolling_mean_3'] = features_df.groupby(group_cols)['Modal_Price'].transform(lambda x: x.rolling(3, min_periods=1).mean())
    features_df['rolling_mean_6'] = features_df.groupby(group_cols)['Modal_Price'].transform(lambda x: x.rolling(6, min_periods=1).mean())

    # Month and year as features
    features_df['month_sin'] = np.sin(2 * np.pi * features_df['Month']/12)
    features_df['month_cos'] = np.cos(2 * np.pi * features_df['Month']/12)

    return features_df

# Apply feature engineering
group_cols = ['STATE', 'District Name', 'Market Name', 'Commodity']
features_df = create_features(monthly_avg, group_cols)

# Encode categorical variables
label_encoders = {}
for col in ['STATE', 'District Name', 'Market Name', 'Commodity']:
    le = LabelEncoder()
    features_df[col + '_encoded'] = le.fit_transform(features_df[col])
    label_encoders[col] = le

# Prepare features and target
feature_cols = ['lag_1', 'lag_2', 'lag_3', 'rolling_mean_3', 'rolling_mean_6',
                'month_sin', 'month_cos', 'STATE_encoded', 'District Name_encoded',
                'Market Name_encoded', 'Commodity_encoded', 'Year', 'Month']

# Filter out rows with missing lag values (first few months)
model_df = features_df.dropna(subset=feature_cols)

# Train-test split (use last 3 months for testing)
latest_dates = model_df['Year-Month'].unique()
latest_dates = pd.Series(latest_dates).sort_values()
test_months = latest_dates[-3:]
train_df = model_df[~model_df['Year-Month'].isin(test_months)]
test_df = model_df[model_df['Year-Month'].isin(test_months)]

print(f"Training data: {train_df.shape[0]} records")
print(f"Testing data: {test_df.shape[0]} records")

# Train Random Forest model
X_train = train_df[feature_cols]
y_train = train_df['Modal_Price']
X_test = test_df[feature_cols]
y_test = test_df['Modal_Price']

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluate model
train_mae = mean_absolute_error(y_train, train_preds)
test_mae = mean_absolute_error(y_test, test_preds)
train_rmse = np.sqrt(mean_squared_error(y_train, train_preds))
test_rmse = np.sqrt(mean_squared_error(y_test, test_preds))

print(f"Train MAE: {train_mae:.2f}, RMSE: {train_rmse:.2f}")
print(f"Test MAE: {test_mae:.2f}, RMSE: {test_rmse:.2f}")

# Save the model and encoders
joblib.dump(model, 'price_forecast_model.pkl')
joblib.dump(label_encoders, 'label_encoders.pkl')
joblib.dump(feature_cols, 'feature_columns.pkl')

print("Model saved successfully!")

# Function to make future predictions
def forecast_future_prices(state, district, market, commodity, months_ahead=6):
    """
    Forecast prices for the next few months
    """
    # Load model and encoders
    model = joblib.load('price_forecast_model.pkl')
    label_encoders = joblib.load('label_encoders.pkl')
    feature_cols = joblib.load('feature_columns.pkl')

    # Get the latest data for this commodity-market combination
    latest_data = features_df[
        (features_df['STATE'] == state) &
        (features_df['District Name'] == district) &
        (features_df['Market Name'] == market) &
        (features_df['Commodity'] == commodity)
    ].sort_values('Year-Month').tail(3)

    if len(latest_data) == 0:
        return None, "No historical data found for this combination"

    # Prepare future predictions
    forecasts = []
    last_date = latest_data.iloc[-1]['Year-Month']
    current_data = latest_data.iloc[-1:].copy()

    for i in range(1, months_ahead + 1):
        # Calculate next month and year
        next_month = last_date.month + i
        next_year = last_date.year
        if next_month > 12:
            next_month -= 12
            next_year += 1

        # Prepare feature vector
        features = current_data[feature_cols].copy()

        # Update features for the next month
        features['Month'] = next_month
        features['Year'] = next_year
        features['month_sin'] = np.sin(2 * np.pi * features['Month']/12)
        features['month_cos'] = np.cos(2 * np.pi * features['Month']/12)

        # Make prediction
        pred_price = model.predict(features)[0]
        forecasts.append({
            'Year': next_year,
            'Month': next_month,
            'Predicted_Price': pred_price
        })

        # Update lag features for next prediction
        current_data['lag_3'] = current_data['lag_2']
        current_data['lag_2'] = current_data['lag_1']
        current_data['lag_1'] = pred_price
        current_data['rolling_mean_3'] = (current_data['lag_1'] + current_data['lag_2'] + current_data['lag_3']) / 3
        current_data['rolling_mean_6'] = (current_data['rolling_mean_6'] * 5 + pred_price) / 6  # Approximate

    return forecasts, "Success"

# Function to recommend best market
def recommend_market(state, district, commodity, months_ahead=3):
    """
    Find the best market to sell in for the given commodity
    """
    # Get all markets in the district
    markets = features_df[
        (features_df['STATE'] == state) &
        (features_df['District Name'] == district) &
        (features_df['Commodity'] == commodity)
    ]['Market Name'].unique()

    if len(markets) == 0:
        return None, "No market data found for this commodity in the district"

    market_forecasts = {}

    for market in markets:
        forecasts, status = forecast_future_prices(state, district, market, commodity, months_ahead)
        if status == "Success":
            # Use the average of next 3 months predictions
            avg_price = np.mean([f['Predicted_Price'] for f in forecasts[:3]])
            market_forecasts[market] = avg_price

    if not market_forecasts:
        return None, "Could not generate forecasts for any markets"

    # Sort markets by predicted price (highest first)
    sorted_markets = sorted(market_forecasts.items(), key=lambda x: x[1], reverse=True)

    # Apply 10% transport cost adjustment
    recommendations = []
    for i, (market, price) in enumerate(sorted_markets):
        adjusted_price = price * 0.9  # Subtract 10% for transport
        recommendations.append({
            'Market': market,
            'Predicted_Price': price,
            'Adjusted_Price': adjusted_price,
            'Rank': i + 1
        })

    return recommendations, "Success"

# Test the functions
state = "Maharashtra"
district = "nashik"
commodity = "Onion"

# Get market recommendations
recommendations, status = recommend_market(state, district, commodity)
if status == "Success":
    print(f"\nMarket recommendations for {commodity} in {district}, {state}:")
    for rec in recommendations[:3]:  # Top 3 recommendations
        print(f"{rec['Rank']}. {rec['Market']}: ₹{rec['Predicted_Price']:.2f} (₹{rec['Adjusted_Price']:.2f} after transport)")

# Get price forecast for a specific market
market = "Lasalgaon(Niphad)"
forecasts, status = forecast_future_prices(state, district, market, commodity)
if status == "Success":
    print(f"\nPrice forecast for {commodity} in {market}:")
    for f in forecasts:
        print(f"{f['Year']}-{f['Month']:02d}: ₹{f['Predicted_Price']:.2f}")

# Create a simple visualization
def plot_forecast(state, district, market, commodity):
    """
    Plot historical prices and forecast
    """
    # Get historical data
    historical = features_df[
        (features_df['STATE'] == state) &
        (features_df['District Name'] == district) &
        (features_df['Market Name'] == market) &
        (features_df['Commodity'] == commodity)
    ]

    # Get forecast
    forecast, status = forecast_future_prices(state, district, market, commodity)

    if status != "Success" or len(historical) == 0:
        print("Not enough data to plot")
        return

    # Prepare data for plotting
    historical['Date'] = historical.apply(lambda x: datetime(x['Year'], x['Month'], 1), axis=1)
    historical = historical.sort_values('Date')

    forecast_dates = [datetime(f['Year'], f['Month'], 1) for f in forecast]
    forecast_prices = [f['Predicted_Price'] for f in forecast]

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(historical['Date'], historical['Modal_Price'], 'b-', label='Historical Prices')
    plt.plot(forecast_dates, forecast_prices, 'r--', label='Forecast')
    plt.xlabel('Date')
    plt.ylabel('Price (₹)')
    plt.title(f'Price Forecast for {commodity} in {market}')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('price_forecast.png')
    plt.show()

# Generate a plot
plot_forecast(state, district, market, commodity)

# Generate farmer recommendations
def generate_recommendation(state, district, commodity):
    """
    Generate farmer-friendly recommendations
    """
    # Get market recommendations
    market_recs, status = recommend_market(state, district, commodity)
    if status != "Success":
        return "Not enough data to generate recommendations"

    # Get price forecast for the best market
    best_market = market_recs[0]['Market']
    forecasts, status = forecast_future_prices(state, district, best_market, commodity, 6)

    if status != "Success":
        return "Not enough data to generate recommendations"

    # Find the best month to sell (highest predicted price)
    best_month = max(forecasts, key=lambda x: x['Predicted_Price'])

    # Current month forecast
    current_forecast = forecasts[0]['Predicted_Price']

    # Generate recommendation
    if best_month['Predicted_Price'] > current_forecast * 1.1:  # At least 10% higher
        recommendation = f"Wait until {best_month['Year']}-{best_month['Month']:02d} to sell at ₹{best_month['Predicted_Price']:.2f}"
    else:
        recommendation = "Sell now for the best available price"

    return f"For {commodity} in {district}, {state}: {recommendation}. Recommended market: {best_market}"

# Generate a recommendation
recommendation = generate_recommendation(state, district, commodity)
print(f"\nFarmer Recommendation:\n{recommendation}")
```
